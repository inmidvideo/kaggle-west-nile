```python
%pylab inline

import pandas as pd
import numpy as np
```

```python
# Read data
train = pd.read_csv('../working/train_f.csv', index_col=0, dtype='float64')
train.shape
```

```python
# add extra data
def gen_data(data, f):
    x = data
    std = data.std()
    
    # iterate through rows
    for index, row in data.iterrows():
        if data.iloc[index]['WnvPresent'] > 0:
            n = int(data.iloc[index]['NumMosquitos'])
            for i in range(n/f):
                #x = x.append(row + f*std*numpy.random.uniform(low=-1.0, high=1.0, size=len(std)), ignore_index=True)
                x = x.append(row, ignore_index=True)
            #break
    
    return x
```

```python
train_d = gen_data(train, 2)
train_d.shape
```
